```python
from PIL import Image
import numpy as np
import pandas as pd
from sklearn import decomposition
from sklearn.svm import SVC
from pylab import * 
from sklearn.cross_validation import train_test_split
from os import listdir
from os.path import isfile, join

DIMENSION = 32
ROOT_DIR = "C:/Users/Dameria Desi/Documents/TA/IMAGE/images/best_seller"
ROOT_DIR2 = "C:/Users/Dameria Desi/Documents/TA/IMAGE/images/non_best_seller"
BEST = "best_seller"
NBEST = "non_best_seller"
CLASSES = [BEST, NBEST]

def getData(generateTuningData):
    trainingData = {}
    tuneData = {}
    testData = {}
    for clazz in CLASSES:
        (train, tune, test) = buildTrainTestVectors(buildImageList(ROOT_DIR + clazz + "/"), generateTuningData)
        trainingData[clazz] = train
        tuneData[clazz] = tune
        testData[clazz] = test

    return (trainingData, tuneData, testData)

def buildImageList(dirName):
    imgs = [Image.open(dirName + fileName).resize((DIMENSION, DIMENSION)) for fileName in os.listdir(dirName)]
    imgs = [list(itertools.chain.from_iterable(img.getdata())) for img in imgs]
    return imgs

Best = [f for f in listdir(ROOT_DIR) if isfile(join(ROOT_DIR, f))]
NBest = [f for f in listdir(ROOT_DIR) if isfile(join(ROOT_DIR2, f))]

for i in Best:
 Image.open(join(ROOT_DIR,i))
 a = Image.open(join(ROOT_DIR,i)) 
matrix_Best = array([array(a).flatten()
                  for i in Best],'f')

pca = decomposition.PCA(n_components=10) # dimensional reduction PCA
pca.fit(matrix_Best)

for i in NBest:
 Image.open(join(ROOT_DIR2,i))
 b = Image.open(join(ROOT_DIR2,i)) 
matrix_NBest = array([array(b).flatten()
                  for i in NBest ],'f')
pca = decomposition.PCA(n_components=10) # dimensional reduction PCA
pca.fit(matrix_NBest)

visual_fiture= []
for i in Best:
    visual_fiture.append(matrix_Best);

for i in Best:
    visual_fiture.append(matrix_NBest);


kategori_list= []
for k in range(len(Best)):
   kategori_list.append("yes" );  

for k in range(len(NBest)):
   kategori_list.append("no");
data = pd.DataFrame([list(visual_fiture),list(kategori_list)])
data = data.T
```
